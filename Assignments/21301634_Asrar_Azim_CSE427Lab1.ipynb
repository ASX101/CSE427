```python
import numpy as np
```

Numpy

```python
#1

arr1 = np.random.randint(1, 101, 20)
arr2 = np.random.randint(1, 101, 20)
print("Array a:\n", arr1)
print("Array b:\n", arr2)
```

```python
#2

a = arr1.reshape(5, 4)
b = arr2.reshape(5, 4)
print("\nMatrix a:\n", a)
print("\nMatrix b:\n", b)
```

```python
#3

scalar_mult_a = a * 2
scalar_mult_b = b * 3
print("Scalar Multiplication:\n", scalar_mult_a)
print("Scalar Multiplication:\n", scalar_mult_b)
```

```python
#4
b_mult = b.T
c = np.matmul(a, b_mult)
print("\nMatrix Multiplication (a * b.T):\n", c)
```

```python
#5

print("\nColumns 2 and 3 of matrix c:\n", c[:, 1:3])
print("\nWhole matrix c using slicing:\n", c[:])
```

```python
#6

max_value = np.max(c)
max_position = np.argmax(c)
max_row, max_col = np.unravel_index(max_position, c.shape)

print(f"\nLargest value in matrix c: {max_value}")
print(f"Index of largest value: ({max_row}, {max_col})")
```

```python
#7

c_flat = c.flatten()
print("\n1D array from matrix c:\n", c_flat)
```

Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
#1


temperature_dhaka = [25, 34, 21, 45, 28, 6, 43, 18, 7, 2]
humidity_dhaka = [28, 25, 29, 20, 26, 50, 19, 29, 52, 55]


plt.scatter(temperature_dhaka, humidity_dhaka, color='red', marker='*')
plt.title('Temperature vs Humidity (Dhaka)')
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.grid(True)
plt.show()
```

```python
#2

study_hours = [2,3,4,4,5,6,7,7,8,9,9,10,11,11,12]
marks = [6,10,15,20,34,44,55,60,55,67,70,80,90,99,100]
plt.figure(figsize=(12,8))
plt.plot(study_hours, marks, marker='o', linestyle='-')
plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

```python
#3

subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks = [89, 90, 45, 78, 99]
plt.barh(subjects, marks, color='green')
plt.title('Subject Wise Scores')
plt.xlabel('Marks')
plt.show()
```

```python
#4

plt.bar(subjects, marks, color=['green', 'red', 'orange', 'purple', 'blue'])
plt.title('Subject Wise Scores (Vertical)')
plt.ylabel('Marks')
plt.show()
```

```python
#5


temperature_dhaka = [25, 34, 21, 45, 28, 6, 43, 18, 7, 2]
humidity_dhaka = [28, 25, 29, 20, 26, 50, 19, 29, 52, 55]
study_hours = [2,3,4,4,5,6,7,7,8,9,9,10,11,11,12]
marks_1 = [6,10,15,20,34,44,55,60,55,67,70,80,90,99,100]
subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks_2 = [89, 90, 45, 78, 99]

plt.figure(figsize=(12, 8))
# 1 scatter
plt.subplot(2, 2, 1)
plt.scatter(temperature_dhaka, humidity_dhaka, color='red', marker='*')
plt.xlabel('Temperature (Dhaka)')
plt.ylabel('Humidity (Dhaka)')
plt.title('Temperature vs Humidity')
# 2 line plot
plt.subplot(2, 2, 2)
plt.plot(study_hours, marks_1, marker='o', linestyle='-', color='red')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title('Study Hours vs Marks')
# 3 horizontal
plt.subplot(2, 2, 3)
plt.barh(subjects, marks_2, color='green')
plt.xlabel('Marks')
plt.ylabel('Subjects')
plt.title('Scores in Each Subject (Horizontal Bar)')
# 4 Vertical
plt.subplot(2, 2, 4)
plt.bar(subjects, marks_2, color=['green', 'red', 'orange', 'purple', 'blue'])
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Scores in Each Subject (Vertical Bar)')
plt.tight_layout()
plt.show()
```
